## How to verify control algorithm?
 Verification of any control algorithm is not trivial task and the best situation is, when it is possible to implement the algorithm on the real life objects.

 In the case of flying robots it is especially important because of the difficulties with modeling of environment, and the object itself. Winds, rain, changes of temperature, pressure etc. have significant impact on the behaviour of the object. It is common knowledge. But even model of flying robot often has important flaws. Mainly because of the characteristics of actuation systems. Forces genereted by propellers or ailerons are heavily dependent on many factors, the modelling of which can complicate the model to the impractical level. It is one of the reasons why wind tunnels are still in use.

 Despite the difficulties described above numerical simulations can be very helpful at the initial stage of the verification. Especially when it is possible to use computational fluid dynamics methods, which are still very computationally demanding.

 The important thing is to remember that it is **necessary to understand
 limitations of the simulations**.

## How to simulate a dynamic object?
 For the purposes of presentation let analyse simple aeroplane which fly only on some fixed altitude. Because of that our simulation will analyse behaviour of the object moving on the plane. Lets call it XY-plane, from the two axis which describes two possible directions. Our plane can also change its orientation, which can be expressed as $\theta$ - angle of rotation between global frame and the frame attached to the plane.

 We also assume that the object can be approximated by a rigid body, and is fully actuated. It can generate propulsion forces ($\tau_x$, $\tau_y$) which acts along X-axis and Y-axis of the local frame, and torque ($\tau_\theta$) which can change orientation of the object.

 Our plane will not fly in vacuum, so we also include effect of drag force proportional to velocity squared which acts equally in each direction.

 To summarize the assumptions:

 1. Movement only on the plane.
 2. Three degrees of freedom (3DOF), movement in X and Y direction, and rotation about axis perpendicular to XY-plane.
 3. Can be approximated by rigid body.

 ![Aeroplane on plane](gfx/aeroplane_on_plane.svg "Aeroplane on XY-plane")

Let define the state vector $\vx$, vector of local velocities $\vn$ and actuation forces vector $\vt$ as:

$\vx = \BM x \\ y \\ \theta \EBM$, $\vn = \BM u \\ v \\ r \EBM$, $\vt = \BM \tau_x \\ \tau_y \\ \tau_\theta \EBM$, 

where $u$ and $v$ are velocities along X-axis, Y-axis of local frame, and $r$ is velocity of rotation. In the case of movement on the 2D plane, there is no difference between the rotation velocity in local and in global frames.

The equation of kinematics can be written as:

$\dot{\vx} = R^b_a \vn = \BM cos(\theta) & -sin(\theta) & 0 \\ sin(\theta) & \cos(\theta) & 0 \\ 0 & 0 & 1 \EBM \vn$

The equation of dynamics can be written as:

$\vt = M \dot{\vn} + C(\vn) \vn + D(\vn)$,

where $M$ is mass matrix, $C$ is centrifugal and Coriolis forces matrix and D is vector describing damping forces.


## Model parameters
Let assume the form of matrices $M$ and $C$ as:

$M = \BM m & 0 & 0 \\ 0 & m & 0 \\ 0 & 0 & I_z \EBM$, $C = \BM 0 & -r \cdot m & 0 \\ r \cdot m & 0 & 0 \\ 0 & 0 & 0 \EBM$, 

where $m = 2$ and $I_z = 0.5$.

The damping forces element $D$ depends on velocity squared, so let assume it form as:

$D = \vec{k_D} \circ sign(\vn) \circ \vn \circ \vn$, 

where $sign \in \{-1,1\}$, and $\circ$ denotes element-wise multiplication (Hadamard product), and $\vec{k_D}$ is damping coefficients vector and is equal to:

$\vec{k_D} = \BM 0.1 \\ 0.5 \\ 2\EBM$.

With these parameters, implement an object simulation.


### Implementation note
In ```numpy``` library:
- element-wise multiplication: $\circ$ $\rightarrow$ ```*```
- matrix multiplication $\rightarrow$ ```@```


````python
# import of necessary libraries
import numpy as np # helpful with matrix operations
from scipy.integrate import solve_ivp # solving differential equations
import matplotlib.pyplot as plt # ploting results


# define shortcuts for a more readable notation
MAT = np.array
````

````python
# define model parameters
m = 2  # mass
Iz = 0.5  # moment of inertia

# mass matrix
M = MAT( [ [m, 0., 0.], 
           [0., m, 0.], 
           [0., 0., Iz] ] )

# C is dependent on r therefore:
def C(vel): # vel = [u v r]
    r = vel[2][0]
    return MAT( [ [0, -r*m, 0],
                  [r*m, 0,  0],
                  [0,   0,  0]] )

# damping coefficients
kD = MAT([ [0.1],
           [0.5],
           [2] ] )  # damping coefficient

# damping forces
def D(vel):
    return kD*np.sign(vel)*vel*vel
````
